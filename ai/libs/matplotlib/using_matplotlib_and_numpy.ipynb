```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

# Criando uma figura b√°sica

```python
# O comando np.arange(in√≠cio, fim, passo) cria um vetor cujo primeiro elemento √© o n√∫mero
# √© o in√≠cio e o √∫ltimo elemento √© o fim, os elementos entre estes dois s√£o espa√ßados de passo em
# passo
t = np.arange(0.01, 5.0, 0.01)

# O comando np.exp() calcula a exponencial do n√∫mero entre par√™nteses. Se o par√¢metro for
# um vetor, ent√£o calcula o valor para cada elemento do vetor.
s = np.exp(-t)

# Instanciando a figura e os eixos
fig, ax = plt.subplots()

# Desenha a figura com base nos pontos t (para o eixo x) e s (para o eixo y)
ax.plot(t, s)

# Altera o label do eixo x
ax.set_xlabel('Tempo')

# Altera o label do eixo y
ax.set_ylabel('Quantidade')

# Altera o t√≠tulo da figura
ax.set_title('Should be growing...')

# Exibe as linhas auxiliares no gr√°fico
ax.grid(True)

# Apresenta a imagem na tela
plt.show()
```

# Criando um Histograma

```python
N_points = 100000
n_bins = 10

rng = np.random.default_rng(1000)
dist = rng.standard_normal(N_points)

fig, ax = plt.subplots()

ax.hist(dist, bins=n_bins)

ax.set_xlabel('N√∫mero de Ocorr√™ncias')
ax.set_ylabel('Valores')
ax.set_title('Histograma')

plt.show()
```
