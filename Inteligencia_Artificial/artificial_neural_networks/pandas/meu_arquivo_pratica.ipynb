### Criando um dataframe a partir de um dicionário

```python
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }

vendas_df = pd.DataFrame(venda)
vendas_df
```

```python
display(vendas_df.head())
print(vendas_df.shape)
display(vendas_df.describe())
```

### Pegar 1 coluna (e os pd.Series)

```python
produtos = vendas_df['Produto'] # Não é um dataframe quando pega uma coluna, e sim uma Série do pandas (pd.Series)
display(produtos)

prod_and_id = vendas_df[['Produto', 'ID Loja']]
display(prod_and_id)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# Uma linha
display(vendas_df.loc[1])
```

```python
# Mais de uma linha
display(vendas_df.loc[1:5])
```

```python
# Pegar linhas que correspondem a uma condição - 1
display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping'])
```

```python
# Pegar linhas que correspondem a uma condição - 2
display(vendas_df.loc[vendas_df['Valor Final'] < 35])
```

```python
# Condição com especificação das colunas
vendas_norte_shop_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norte_shop_df)
```

```python
# Condição com especificação das colunas
vendas_norte_shop_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norte_shop_df)
```

```python
# Pegar um valor específico
print(vendas_df.loc[1, ['Produto']])
```

### Adicionar 1 coluna

```python
# A partir de uma colune existente
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# Criar coluna com um valor padrão - funciona mas não é a maneira correta
vendas_df['Imposto'] = 0
display(vendas_df)
```

```python
# Criar coluna com um valor padrão - Maneira correta
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
display(vendas_dez_df)
```

```python
# Concatenar dois dataframes e pegar o resultado mesclado em um dataframe
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

>Se atentar aos valores da coluna Comissão e Impostos. Como foram criadas durante o processo, seus valores são NaN. Para isso, pode-se adicionar as colunas no excel, ou >executar um processamento dos dados destas 2 colunas para definir um valor.

### Excluir linhas e colunas

```python
# No caso da coluna à ser removida, o eixo é 1 - Removendo coluna Imposto
vendas_df = vendas_df.drop('Imposto', axis=1)
display(vendas_df)
```

```python
# No caso da linha à ser removida, o eixo é 0 - Removendo linha 0
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# Deletar linhas e colunas completamente vazias

# Deleta todas as linhas vazias - Por padrão as linhas são removidas
vendas_df = vendas_df.dropna(how='all')
# OU
vendas_df = vendas_df.dropna(how='all', axis=0)
display(vendas_df)
```

```python
# Deleta todas as colunas vazias definindo axis=1
vendas_df = vendas_df.dropna(how='all', axis=1)

display(vendas_df)
```

```python
# Deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()
display(vendas_df)
```

```python
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

```python
# Preencher valores da coluna que possui valor vazio pela méda dos valores de comissão
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)
```

```python
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

```python
# Outra maneira de preencher valores de colunas com valores vazios - Preencher valores baseado no último valor preenchido acima
vendas_df['Comissão'] = vendas_df['Comissão'].ffill()
display(vendas_df)
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# Value Counts
transacoes_por_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_por_loja)
```

```python
# Group By - Mas com todas as colunas somadas
total_valor_vendas_produtos = vendas_df.groupby('Produto').sum(numeric_only=True)
display(total_valor_vendas_produtos)
```

```python
# Group By - Mas com apenas as colunas produtos e valor final - Soma total
total_valor_vendas_produtos = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(total_valor_vendas_produtos)
```

```python
# Group By - Mas com apenas as colunas produtos e valor final - Média
total_valor_vendas_produtos = vendas_df[['Produto', 'Valor Final']].groupby('Produto').mean()
display(total_valor_vendas_produtos)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
# Adicionar uma coluna gerente para vendas_df, executando merge com condicional para atrelar o gerente baseado no ID da Loja dos 2 dataframes
# O pandas consegue automaticamente fazer o merge e realizar a condição, já que as colunas possuem o mesmo nome, o Pandas já faz a correlação automaticamente
vendas_df = pd.merge(vendas_df, gerentes_df)
```
