# Problema OneMax

* Busca-se maximizar o número de "1s" durante a busca;
* Inicialmente as soluções são geradas aleatoriamente por uma lista de bits 0 e 1;
* O objetivo é gerar a solução com a maior quantidade de “1”
    * Exemplo:
    * 10010 (soma == 2)
    * 01110 (soma == 3)
    * 11111 (soma == 5)

#### Instalação do framework

```python
# Configurações para execução do framework
import six
import sys
sys.modules['sklearn.externals.six'] = six
import mlrose
```

### Construção da função fitness

```python
# objective function
def one_max_fitness_function(individual):
    """The sum of all ones in the individual (list)."""
    count = 0
    for bit in individual:
        count += bit

    return count
```

### Construção do problema no mlrose

```python
fitness = mlrose.CustomFitness(one_max_fitness_function)
```

```python
problema = mlrose.DiscreteOpt(length = 5, fitness_fn = fitness,
                             maximize = True, max_val = 2)
```

### Algoritmo Genético

```python
solucao_GA, melhor_custo = mlrose.genetic_alg(problema, pop_size=10, mutation_prob=0.2,
                                       max_iters=10)
solucao_GA, melhor_custo
```
