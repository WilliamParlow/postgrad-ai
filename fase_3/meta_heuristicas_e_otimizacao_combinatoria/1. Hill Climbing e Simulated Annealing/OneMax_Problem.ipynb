# Problema OneMax

* Busca-se maximizar o número de "1s" durante a busca; 
* Inicialmente as soluções são geradas aleatoriamente por uma lista de bits 0 e 1;
* O objetivo é gerar a solução com a maior quantidade de “1”
    * Exemplo:
    * 10010 (soma == 2)
    * 01110 (soma == 3)
    * 11111 (soma == 5)

#### Configuração do framework

```python
# Configurações para execução do framework
import six
import sys
sys.modules['sklearn.externals.six'] = six
import mlrose
```

### Construção da função fitness

```python
def fitness_function(solucao):
    """A soma dos valores 1s da solucao (list)."""

    soma = 0

    for bit in solucao:
        soma += bit

    return soma
```

### Construção do problema no mlrose

```python
fitness = mlrose.CustomFitness(fitness_function)
```

```python
problema = mlrose.DiscreteOpt(length = 64, fitness_fn = fitness,
                             maximize = True, max_val = 2)
```

### Algoritmo Hill climbing

```python
solucao_HC, melhor_custo, fitness_curve = mlrose.hill_climb(problema, curve=True)
solucao_HC, melhor_custo, fitness_curve
```

```python
import matplotlib.pyplot as plt

plt.plot(fitness_curve)
plt.title('Fitness Curve - Hill Climbing')
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.show()
```

### Algoritmo Simulated annealing

```python
solucao_SA, melhor_custo, fitness_curve = mlrose.simulated_annealing(problema, curve=True)
solucao_SA, melhor_custo, fitness_curve
```

```python
import matplotlib.pyplot as plt

plt.plot(fitness_curve)
plt.title('Fitness Curve - Simulated Annealing')
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.show()
```
