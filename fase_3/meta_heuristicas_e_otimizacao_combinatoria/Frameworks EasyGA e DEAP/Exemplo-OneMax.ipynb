```python
import EasyGA
import random
```

```python
# Create the Genetic algorithm
ga = EasyGA.GA()

# Chromosome params
ga.population_size = 10
ga.chromosome_length = 20
ga.generation_goal = 50

# Create random genes from 0 to 1
ga.gene_impl = lambda: random.randint(0, 1)

# Minimize the sum of the genes
ga.target_fitness_type = 'max'

# Set the fitness goal
ga.fitness_goal = ga.chromosome_length
```

```python
# def fitnessFunction_BKP(chromosome):
#     # Overall fitness value
#     fitness = 0
#     # For each gene in the chromosome
#     for gene in chromosome.gene_list:
#         # Check if its value = 1
#         if(gene.value == 1):
#             # If its value is 1 then add one to
#             # the overal fitness of the chromosome.
#             fitness += 1

#     return fitness
```

```python
def fitnessFunction(chromosome):
    # Overall fitness value
    fitness = 0
    # For each gene in the chromosome
    for gene in chromosome:
        # Check if its value = 1
        if(gene == 1):
            # If its value is 1 then add one to
            # the overal fitness of the chromosome.
            fitness += 1

    return fitness
```

```python
# Cadastra a função fitness
ga.fitness_function_impl = fitnessFunction
```

```python
ga.evolve()
ga.print_generation()
ga.print_population()
```

```python
# Mostra a curva de convergência
ga.graph.highest_value_chromosome()
ga.graph.show()
```
