```python
# Função objetivo: número de pares de rainhas não-atacadas
# nQueens: Numero de rainhas
# Solução ótima: (nQueens x nQueens-1)/2
def fitnessFunction(solucao):
      melhorFitness = (len(solucao)*(len(solucao)-1)) // 2
      #  Contagem de pares de ataques
      h = 0
      # Verificação dos ataques pela diagonal
      for i in range(0, len(solucao)):
        for j in range(i+1, len(solucao)):
          if abs(i -j) == abs(solucao[i] - solucao[j]):
            h+=1
          # Verificação de ataques na mesma linha
          if  abs(solucao[i] - solucao[j]) == 0:
            h+=1
      return  melhorFitness - h
```

#### Exemplos:
- Solução: [2,7,3,4,6,5,8,1] - Pares não atacados: **22**
- Solução: [2,4,7,4,8,5,5,2] - Pares não atacados: **24**
- Solução: [3,2,5,4,3,2,1,3] - Pares não atacados: **11**

```python
# Testando a função fitness
print(fitnessFunction([2,7,3,4,6,5,8,1]))
print(fitnessFunction([2,4,7,4,8,5,5,2]))
print(fitnessFunction([3,2,5,4,3,2,1,3]))
```
