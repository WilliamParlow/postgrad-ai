1. Execute o código a seguir para fazer o download do dataset para ser utilizado nesses exercícios. O dataset vai ser salvo no sistema de arquivos do ambiente de execução do Colab.

```python
import requests, os

url = "https://raw.githubusercontent.com/watinha/nlp-text-mining-datasets/main/b2w-50.csv"

def download (url):
  if (os.path.isfile('./b2w-50.csv')):
    print('Arquivo já existente no Runtime... Tudo OK')
    return
  response = requests.get(url)
  with open('./b2w-50.csv', 'wb') as f:
      f.write(response.content)
      print('Download realizado e arquivo extraído no Runtime... Tudo OK')

download(url)
```

2. Faça o carregamento do dataset, extraia o corpus do atributo review_text do dataset e contabilize as seguintes propriedades do texto, com a API CountVectorizer:
* Palavras que aparecem mais/menos vezes no texto;
* Quantidade média de palavras por documento.

**Resposta:**
A palavra que aparece mais vezes é a palavra produto (34 vezes).
A palavra que aparece menos vezes é o termo 02 (1 vez).
Os documentos tem em média 16,32 termos.

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer


df = pd.read_csv('./b2w-50.csv')

corpus = df['review_text']

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
vocabulary = vectorizer.get_feature_names_out()

X_df = pd.DataFrame(X.toarray(), columns=vocabulary)
word_count = X_df.sum(axis=0)

max_ind = word_count.argmax()
min_ind = word_count.argmin()

print(f'{vocabulary[max_ind]} aparece mais vezes ({word_count[max_ind]})')
print(f'{vocabulary[min_ind]} aparece menos vezes ({word_count[min_ind]})')

words_average = X_df.sum(axis=1).mean()
print(f'Média de palavras por documento: {words_average}')
```

3. Calcule o IDF do corpus e identifique os 5 termos com maior/menor valor no corpus.

**Resposta:** os termos 02, nao, mão, mãe e máquina são os termos com maior IDF. E os termos produto, de, muito, com entrega são os termos com menor IDF.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(corpus)

idfs = pd.Series(vectorizer.idf_, index=vectorizer.get_feature_names_out())

print(idfs.sort_values(ascending=False).head(5))
print(idfs.sort_values().head(5))
```
