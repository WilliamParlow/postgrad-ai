1. Implemente uma função que remova as stop-words de um texto passado como parâmetro. Lembrem de:

* Atualizar a biblioteca do spacy;
* Carregar o modelo do spacy para a língua portuguesa.

Para verificar se o código deu certo, testem o código com o seguinte texto:

"A Nintendo anunciou o desenvolvimento do jogo com um
teaser trailer na Nintendo Direct da E3 2019. Em fevereiro de 2021,
o produtor da série Zelda, Eiji Aonuma, afirmou que mais informações
sobre a sequência seriam divulgadas no final daquele ano. Na
Nintendo Direct da E3 2021, a Nintendo mostrou um trailer revelando
a jogabilidade e a história, e anunciou a data de lançamento para
2022. Quando o trailer não apresentou um título oficial ou um
close-up do rosto do protagonista, várias teorias de fãs surgiram
sobre o enredo do próximo jogo."

O resultado deve ser o seguinte:

"Nintendo anunciou desenvolvimento jogo
 teaser trailer Nintendo Direct E3 2019 . fevereiro 2021 ,
 produtor série Zelda , Eiji Aonuma , afirmou informações
 sequência seriam divulgadas ano .
 Nintendo Direct E3 2021 , Nintendo mostrou trailer revelando
 jogabilidade história , anunciou data lançamento
 2022 . trailer apresentou título oficial
 close-up rosto protagonista , várias teorias fãs surgiram
 enredo jogo ."

```sh
!pip install -U spacy
!python -m spacy download pt_core_news_sm
```

````python
import spacy
nlp = spacy.load('pt_core_news_sm')

def remove_stop_words(text):
  doc = nlp(text)
  return ' '.join([token.text for token in doc if not token.is_stop])


text = '''A Nintendo anunciou o desenvolvimento do jogo com um
teaser trailer na Nintendo Direct da E3 2019. Em fevereiro de 2021,
o produtor da série Zelda, Eiji Aonuma, afirmou que mais informações
sobre a sequência seriam divulgadas no final daquele ano. Na
Nintendo Direct da E3 2021, a Nintendo mostrou um trailer revelando
a jogabilidade e a história, e anunciou a data de lançamento para
2022. Quando o trailer não apresentou um título oficial ou um
close-up do rosto do protagonista, várias teorias de fãs surgiram
sobre o enredo do próximo jogo.'''

print(remove_stop_words(text))
````

2. Implemente uma função que substitua os termos de uma frase com seus lemas, utilizando a biblioteca Spacy.

Para testar essa função, utilize a saída da função anterior. O resultado deve ser a seguinte string:

"Nintendo anunciar desenvolvimento jogo
  teaser trailer Nintendo Direct E3 2019 . fevereiro 2021 ,
  produtor série Zelda , Eiji Aonuma , afirmar informação
  sequência divulgar ano .
  Nintendo Direct E3 2021 , Nintendo mostrar trailer revelar
  jogabilidade história , anunciar data lançamento
  2022 . trailer apresentar título oficial
  close-up rosto protagonista , várias teoria fã surgir
  enredo jogo ."

````python
def replace_terms_with_lemmas(text):
  doc = nlp(text)
  return ' '.join([token.lemma_ for token in doc])

print(remove_stop_words(replace_terms_with_lemmas(text)))
````

3. Implemente uma função que, dado um texto, retorne as entidades nomeadas que são do tipo: organização (ORG) e lugar (LOC).

Para testar essa função, utilize o texto inicial dos exercícios complementares. O resultado deve ser um vetor contendo os textos associados às entidades nomeadas, com os seguintes elementos:

{'Nintendo'}

````python
def get_entities(text):
  doc = nlp(text)
  return set([ent.text for ent in doc.ents if ent.label_ in ['ORG', 'LOC']])

print(get_entities(text))
````
