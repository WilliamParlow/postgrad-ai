#Embeddings
Embeddings são representações numéricas de objetos do mundo real, como palavras, sentenças, imagens e outros dados, que modelos de aprendizado de máquina usam para entender relações complexas e realizar várias tarefas.

Embeddings são amplamente utilizados no Processamento de Linguagem Natural (PLN) para representar palavras e frases como vetores numéricos em um espaço de alta dimensão. Essa técnica permite que as palavras sejam interpretadas por algoritmos de aprendizado de máquina, facilitando a compreensão e o processamento da linguagem humana por computadores.

Embeddings de palavras podem ser treinados utilizando diferentes estratégias (Glove, Skip-gram e CBOW). Seu processo de treinamento é bastante intensivo computacionalmente e exige um córpus significativo.

##Carregando os Embeddings
Para carregar os Embeddings, vamos utilizar a biblioteca [Gensim](https://radimrehurek.com/gensim/). Mais especificamente vamos utilizar a classe models.KeyedVectors e o método load_word2vec_format.

```sh
pip install gensim
```

````python
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format('./skip_50d.txt')
````

A partir do modelo carregado, nós podemos recuperar Embeddings (vetores) que representam o significado de diferentes palavras.

````python
rei = model.get_vector('rei')
homem = model.get_vector('homem')
mulher = model.get_vector('mulher')

print(rei)
print(homem)
print(mulher)

print(rei.shape)
print(homem.shape)
print(mulher.shape)
````

Nesse exemplo, nós carregamos os Embeddings referentes a 3 palavras: "rei", "homem" e "mulher". Observem que cada um desses Embeddings é representado como um vetor numérico (Numpy array). Esse vetor contém valores que representam o significado da palavra no espaço e podem ser utilizados para a realização de diferentes tarefas de PLN.

É importante observar que todos esse vetores foram representados em vetores com 50 posições (50 dimensões). No website do NILC, 50 dimensões são as Embeddings mais simples, que contém menos informações sobre as palavras. O website disponibiliza Embeddings até 1000 dimensões.

##Calculando similaridade entre palavras
No modelo BOW, as palavras são representadas em documentos, sem considerar a sua semântica. Se tivermos documentos com a palavra "elegante" e "chique", o modelo BOW consideraria que essas duas palavras são completamente diferentes. No entanto, semanticamente, elas podem estar fazendo referência a um conceito similar.

Embeddings são utilizadas para tentar representar essa semântica entre os termos, sendo possível calcular a similaridade entre termos.



````python
def print_similarity(word1, word2):
  similarity = model.similarity(word1, word2)
  print(f'{word1:15} x {word2:15}: {similarity:3}')


print_similarity('elegante', 'avestruz')
print_similarity('elegante', 'feio')
print_similarity('elegante', 'bonito')
print_similarity('elegante', 'chique')
print_similarity('elegante', 'elegante')
````

````python
print_similarity('gato', 'nadadeira')
print_similarity('gato', 'baleia')
print_similarity('gato', 'dinossauro')
print_similarity('gato', 'cachorro')
print_similarity('gato', 'gato')
````

O modelo de Embeddings permite que a realização de buscas por termos similares.

````python
def most_similar(w1):
  similar = model.most_similar(w1)[0:5]
  return '\n - '.join([ f'{s[0]}: {s[1]}' for s in similar ])

print(most_similar('elegante'))
print(most_similar('gato'))
print(most_similar('nadadeiras'))
````

A representação vetorial da semântica dos termos nos Embeddings permite inclusive a realização de operações aritméticas entre termos.

````python
similar = model.most_similar(positive=['rei', 'mulher'], negative=['homem'])
result = [ f' - {str(s)}' for s in similar[0:10]]
print('\n'.join(result))
````

Embeddings também podem ser utilizados para calcular a distância semântica entre diferentes documentos, com o algoritmo Word Mover's Distance. O algoritmo Word Mover's Distance (WMD) é uma técnica para medir a distância semântica entre dois documentos usando modelos de linguagem pré-treinados, como o Word2Vec.

````python
doc1 = 'O nadador saiu da piscina rapidamente.'
print(model.wmdistance(doc1, 'Caiu e levantou do buraco raso.'))
print(model.wmdistance(doc1, 'João jogou a bola para cima.'))
print(model.wmdistance(doc1, 'Maria correu lentamente para casa.'))
print(model.wmdistance(doc1, 'João nadou e saiu da piscina.'))
print(model.wmdistance(doc1, 'O nadador riu e saiu.'))
print(model.wmdistance(doc1, 'O nadador deixou a piscina de forma rápida.'))
````

````python
doc1 = 'O jogador driblou os atacantes e fez o gol.'
print(model.wmdistance(doc1, 'O gato miou e foi para sua casa.'))
print(model.wmdistance(doc1, 'Simões gosta de lagostas.'))
print(model.wmdistance(doc1, 'O jogador fez cena e caiu.'))
print(model.wmdistance(doc1, 'O futebolista fez seu jogo e saiu vencedor.'))
print(model.wmdistance(doc1, 'O jogador pedalou e marcou mais um tento.'))
````
