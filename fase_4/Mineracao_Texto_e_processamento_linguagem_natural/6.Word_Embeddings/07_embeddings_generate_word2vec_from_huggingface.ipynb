#Script para gerar o arquivo de embeddings a partir da biblioteca huggingface

Para executar esse código, vocês vão precisar gerar um token de segredo para permitir o acesso à plataforma HuggingFace a partir do Google Colab.

```python
from huggingface_hub import hf_hub_download
from safetensors.numpy import load_file

path = hf_hub_download(repo_id="nilc-nlp/word2vec-skip-gram-50d",
                       filename="embeddings.safetensors")

data = load_file(path)
vectors = data["embeddings"]

vocab_path = hf_hub_download(repo_id="nilc-nlp/word2vec-skip-gram-50d",
                             filename="vocab.txt")

with open(vocab_path, "r", encoding="utf-8") as f:
    vocab = [line.strip() for line in f if line.strip()]

vocab_size = min(len(vocab), len(vectors))
embedding_dim = vectors.shape[1]

with open("skip_50d.txt", "w", encoding="utf-8") as out:
    out.write(f"{vocab_size} {embedding_dim}\n")
    for word, vector in zip(vocab, vectors):
        values = " ".join(str(v) for v in vector.tolist())
        out.write(f"{word} {values}\n")
```

Esse script vai gerar o arquivo skip_50d.txt que contém as definições de embeddings que podem ser importadas pelo Gensim.
