# Revisão de Aprendizado Supervisionado de Máquina
Nessa semana, vamos estudar sobre Classificação de Documentos Textuais. A Classificação de Documentos utiliza técnicas de Aprendizado Supervisionado com dados extraídos de textos. A Classificação de Documentos Textuais é frequentemente utilizada para realizar tarefas relacionadas à:
*   Análise de Sentimentos em Postagens e Comentários;
*   Detectores de Spam em E-Mails;
*   Entre outras tarefas.

Para analisarmos como implementar essas tarefas, vamos, inicialmente, revisar alguns conceitos sobre Aprendizado de Máquina e sobre o uso do Scikit Learning para implementar scripts de classificação em geral.

Aprendizado de Máquina é um campo da inteligência artificial que se concentra no desenvolvimento de algoritmos e modelos que permitem aos computadores aprender padrões e tomar decisões com base em dados, sem serem explicitamente programados. Essa abordagem permite que os sistemas aprendam com experiências passadas e melhorem seu desempenho ao longo do tempo, sendo amplamente utilizado em uma variedade de aplicações, como reconhecimento de padrões, previsão, classificação e otimização.

O Aprendizado Supervisionado é uma abordagem do Aprendizado de Máquina em que os modelos são treinados em um conjunto de dados rotulados, ou seja, dados em que as saídas desejadas já são conhecidas. Durante o treinamento, o modelo aprende a mapear as entradas para as saídas corretas, permitindo que generalize e faça previsões precisas em novos dados.

A seguir, temos um exemplo de código que carrega do iris dataset que será utilizado para exemplificar conceitos fundamentais de Aprendizado Supervisionado.

```python
import pandas as pd
from sklearn import datasets

dataset = datasets.load_iris()
features_df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(features_df)
```

Nesse código, é carregado o iris dataset a partir da biblioteca sklearn. O [iris dataset](https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html) é frequentemente utilizado como exemplo em aulas sobre aprendizado de máquina e é disponibilizado de forma facilitada na biblioteca sklearn. Esse dataset apresenta caracterísiticas de 3 tipos de flores e o objetivo é utilizar estratégias de aprendizado supervisionado para prever o tipo de flor a partir das características apresentadas.

A partir do iris dataset, são extraídas as características (features) do nosso problema de classificação: sepal length, sepal width, petal length e petal width. Esse problema de classificação é caracterizado por ser um problema com 4 características em um dataset com 150 tuplas (linhas). Para cada conjunto de características, esse dataset também disponibiliza uma classe/tipo de flor associada, codificada como um valor inteiro (0 para setosa, 1 para versicolor e 2 para virginica).

As características, juntamente com os rótulos/classes/tipos de flores devem ser utilizados para treinar um algoritmos de Aprendizado Supervisionado para induzir um modelo preditor que seja capaz de identificar o tipo de flor, a partir dos dados de treinamento. O script a seguir exemplifica a configuração das features, juntamente com os rótulos para treinar um modelo de predição com o algoritmo de Árvore de Decisão.

```python
from sklearn.tree import DecisionTreeClassifier

X = features_df.to_numpy()
y = dataset.target

model = DecisionTreeClassifier()
model.fit(X, y)
```

Existem diferentes modelos de classificação que poderiam ser utilizados nesse mesmo cenário. O sklearn implementa diferentes modelos de Aprendizado Supervisionado, todos com a mesma interface de uso Predictor, como por exemplo:
*   Support Vector Machines - SVM ([SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC));
*   Naive Bayes ([NB](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html#sklearn.naive_bayes.GaussianNB));
*   Random Forest ([RF](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html#sklearn.ensemble.RandomForestClassifier));
*   Entre outros.

A partir do modelo gerado, é possível fazer predições de tipos de flores a partir de um determinado conjunto de características.

```python
X_test = [
    [5, 3, 1, 0.2], # Flor 1
    [6, 2, 5, 2], # Flor 2
    [7, 3, 5, 2] # Flor 3
]

y_pred = model.predict(X_test)
print([dataset.target_names[i] for i in y_pred])
```

# Validação e Teste de Modelos
Nem sempre é fácil gerar dados para testar e avaliar a qualidade dos preditores treinados. Nesse contexto, existem diferentes abordagens que podem ser utilizadas para separar um dataset com os rótulos em datasets de treinamento e teste.

Uma das abordagens mais simples de serem utilizadas pode ser implementada com o método [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html#sklearn.model_selection.train_test_split) do sklearn.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

O método train_test_split separa um dataset em dois grupos: um de treinamento e um de teste. A idéia é utilizar esse método para avaliar um preditor, treinando um modelo de classificação com o grupo de treinamento e testando esse modelo com o grupo de teste. Nesse exemplo, o grupo de treinamento possui 112 amostrar, enquanto o grupo de teste possui 38.

O intuito de usar essa abordagem é garantir que os dados utilizados no teste não tenham sido utilizados no treinamento.

Existem diferentes métricas que podem ser utilizadas para avaliar a performance dos preditores, como F-Score, Precision, Recall, Accuracy, entre outras. O sklearn implementa diferentes formas de se calcular essas métricas ([metrics](https://scikit-learn.org/stable/modules/model_evaluation.html#classification-metrics). No exemplo a seguir, utilizaremos o método [classification_report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html#sklearn.metrics.classification_report) que apresenta um sumário de acurácia com base nas métricas de Acurácia, F-Score, Precision e Recall.

```python
from sklearn.metrics import classification_report

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

# Classificação de Documentos utilizando o modelo Bag-Of-Words (BOW)
O modelo Bag-Of-Words (BOW) trata-se de uma representação de computacional para textos e é frequentemente utilizado para implementação de tarefas associadas à Classificação de Documentos. Nesse contexto, o BOW é utilizado como uma estratégia de extração de características (Feature Extraction), gerando um vetor de valores numéricos (vetor de características) para um documento textual que pode ser usado para treinar algorítmos de Aprendizado Máquina.

Uma das possíveis aplicações para essa abordagem é a Classificação de Sentimentos. A tarefa de Classificação de Sentimentos, também conhecida como Análise de Sentimentos, é uma área da Ciência de Dados e Processamento de Linguagem Natural (PLN) que visa identificar e categorizar as emoções expressas em um determinado texto. Isso pode incluir identificar se um texto é positivo, negativo ou neutro, ou até mesmo identificar emoções específicas, como alegria, tristeza, raiva, entre outras.

A Análise de Sentimentos é utilizada por empresas para quantificar o número de clientes satisfeitos com um produto/serviço a partir de comentários realizados em redes sociais, por exemplo. Na plataforma Kaggle, existem diferentes datasets que dispõem de dados sobre análise de sentimentos de postagens/comentários em diferentes plataformas. Para exemplificar esse caso de uso da tarefa de Classificação de Documentos, vamos utilizar um dataset de comentários da plataforma Busca pé ([Dataset de Análise de Sentimentos no Kaggle](https://www.kaggle.com/datasets/fredericods/ptbr-sentiment-analysis-datasets)).

O código a seguir, faz o download desse dataset e o armazena na plataforma Colab. Vocês podem utilizar esse código em suas próprias máquinas ou simplesmente fazer o download e descompressão do arquivo manualmente em suas máquinas, se preferirem.

```python
import requests, os

url = "https://raw.githubusercontent.com/watinha/nlp-text-mining-datasets/main/buscape.csv"

def download (url):
  if (os.path.isfile('./buscape.csv')):
    print('Arquivo já existente no Runtime... Tudo OK')
    return
  response = requests.get(url)
  with open('./buscape.csv', 'wb') as f:
      f.write(response.content)
      print('Download realizado e arquivo extraído no Runtime... Tudo OK')

download(url)
```

# Características do Dataset
A seguir, carregar o dataset disponibilizado em formato CSV com a API do Pandas. No data frame carregado, observem as colunas:
* review_text que representa o texto de um comentário para um determinado produto;
* polarity que representa se o comentário foi positivo (1,0) ou negativo (0,0).

O dataset completo possui 84.991 linhas, sendo que existem 66.817 comentários positivos e 6.810 comentários negativos.

```python
import pandas as pd

df = pd.read_csv('buscape.csv')
print(df[['review_text', 'polarity']])

print(df['polarity'].value_counts())
```

É importante observar que esse dataset possui alguns valores que podem gerar problemas para a implementação do nosso Classificador de Sentimentos. Na coluna de polaridade, existem valores como NaN (not a number) que precisam ser filtrados desses dados. Esse tipo de valor também pode ser observado na coluna de review_text. Por isso, precisamos remover esses valores dos nossos dados.

```python
import numpy as np

# PREPROCESSAMENTO
# removendo polaridades que sejam NaN
# e garantir que os valores dos comentários não sejam NaN
df_filtered = df.dropna()
```

# Treinando o modelo de Classificação de Sentimentos
Para modelar o problema de classificação de documentos, vamos realizar os seguintes passos:
1.   Extrair os textos em um vetor de strings (corpus) e as polaridades em um vetor de rótulos (y);
2.   Vamos gerar os grupos de treinamento e teste com o método train_test_split;
3.   Extrair as características a partir dos textos do corpus de treinamento, gerando representações BOW para cada comentário do dataset de treinamento (com a API CountVectorizer);
4.   Treinar um modelo de Aprendizado Supervisionado com o vetor de características (X) gerado pelo extrator de características (CountVectorizer);

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

corpus = df_filtered['review_text'].tolist()                             # (1)
y = df_filtered['polarity'].tolist()

corpus_train, corpus_test, y_train, y_test = train_test_split(corpus, y) # (2)

extractor = CountVectorizer(binary=True)                                 # (3)
X_train = extractor.fit_transform(corpus_train)

model = DecisionTreeClassifier()                                         # (4)
model.fit(X_train, y_train)
```

# Avaliando o Modelo de Classificação de Sentimentos
Para avaliar o modelo de aprendizado gerado, devem ser realizados os seguintes passos:
1.   Extrair as características a partir dos textos do corpus de teste;
2.   Testar a acurácia do modelo gerado com o grupo de teste.


```python
X_test = extractor.transform(corpus_test)                                # (1)

y_pred = model.predict(X_test)                                           # (2)
print(classification_report(y_test, y_pred))
```

# TfidfVectorizer
O TfidfVectorizer é uma técnica de extração de características utilizada para representar documentos textuais em um formato numérico, levando em consideração a importância relativa de cada palavra no contexto do documento e do corpus. O TfidfVectorizer é uma extensão do CountVectorizer, que simplesmente conta a frequência de palavras em um documento, mas o TfidfVectorizer vai além, atribuindo pesos às palavras com base na frequência inversa do documento (IDF - Inverse Document Frequency).

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus_train, corpus_test, y_train, y_test = train_test_split(corpus, y) # (2)

extractor = TfidfVectorizer()                                 # (3)
X_train = extractor.fit_transform(corpus_train)

model = DecisionTreeClassifier()                                         # (4)
model.fit(X_train, y_train)
```

```python
X_test = extractor.transform(corpus_test)                                # (1)

y_pred = model.predict(X_test)                                           # (2)
print(classification_report(y_test, y_pred))
```

Em termos mais simples, é a quantidade de informações que estão sendo usadas para distinguir entre diferentes classes ou categorias. Quanto maior a dimensionalidade, mais complexo é o problema de classificação.
