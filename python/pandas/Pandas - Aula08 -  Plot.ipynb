# Exibição do gráfico com pyplot

Para exibir um gráfico usando pyplot, pode-se invocar o método **plot** do DataFrame

```python
df.plot()
```

# Gráfico de dispersão (Scatter)

O método **plot** pode exibir um gráfico de dispersão (scatter). Para isso, é preciso passar 3 argumentos para o método **plot**:
kind = 'scatter'
x = <coluna_A>
y = <coluna_B>

```python
df.plot(kind='scatter', x='Duration', y='Calories')

df.corr()
```

```python
df.plot(kind='scatter', x='Pulse', y='Maxpulse')
```

O gráfico acima mostra que há o relacionamento entre as colunas, de modo que quanto maior a duração do exercício, maior as calorias gastas. Isso corrobora com a tabela de correlação da aula anterior em que as colunas 'Duration' e 'Calories' apresentaram uma taxa de 0.92 de correlação.

No gráfico abaixo, iremos plotar um gráfico de dispersão das colunas 'Duration' e 'Maxpulse', lembrando que a taxa de correlação delas foi de 0.009.

```python
df.plot(kind='scatter', x='Duration', y='Maxpulse')
```

# Histograma

Um histograma é um gráfico que apresenta a frequência de um valor em uma coluna classificado em intervalos.

Para plotar um histograma com o método **plot** é preciso invocar o método a partir de uma coluna do DataFrame passando o argumento **kind='hist'**.

```python
df['Duration'].plot(kind='hist', bins=10, rwidth=0.95, title='Duration Histogram - Modo 1')
```

```python
df.hist(column='Duration', grid=False, rwidth=.95)
```
