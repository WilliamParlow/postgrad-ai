# Lists

```python
# length of the list
print(len([1, 2, 3, 4, 5]))

# List access by negative index
print([1, 2, 3, 4, 5][-1])

# List slicing
print([1, 2, 3, 4, 5][1:3])

# List slicing with step
print([1, 2, 3, 4, 5][::2])

# List slicing with negative step
print([1, 2, 3, 4, 5][::-1])

# List slicing with start and end
print([1, 2, 3, 4, 5][:4])

# List slicing with start and end
print([1, 2, 3, 4, 5][2:])

# List slicing with start and end and step
print([1, 2, 3, 4, 5][1:4:2])

# List slicing with start and end and negative step
print([1, 2, 3, 4, 5][4:1:-1])

# check if value is in list
print(1 in [1, 2, 3, 4, 5])

# Also can be used with if
if 1 in [1, 2, 3, 4, 5]:
    print("1 is in the list")
    
# modify list with intervals
list1 = [1, 2, 3, 4, 5]
list1[1:3] = [6, 7]
print(list1)  # Output: [1, 6, 7, 4, 5]

# modify list with intervals and step
list1 = [1, 2, 3, 4, 5]
list1[::3] = [6, 7]
print(list1)  # Output: [6, 2, 7, 4, 5]

# modify list with intervals and negative step
list1 = [1, 2, 3, 4, 5]
list1[::-1] = [6, 7, 8, 9, 10]
print(list1)  # Output: [10, 9, 8, 7, 6]

list1 = [5 , 10 , 15 , 20 , 30 , 40 , 50 , 60 , 70 , 80]
list1[2:5] = [1000] * (5 - 2)
print(list1)  # Output: [5, 10, 1000, 30, 40, 50, 60, 70, 80]

# add new element
list1 = [1, 2, 3, 4, 5]
list1.append(6)
print(list1)  # Output: [1, 2, 3, 4, 5, 6]

# add new element at the beginning
list1 = [1, 2, 3, 4, 5]
list1.insert(0, 0)
print(list1)  # Output: [0, 1, 2, 3, 4, 5]

# extend list
list1 = [1, 2, 3, 4, 5]
list1.extend([6, 7, 8])
print(list1)  # Output: [1, 2, 3, 4, 5, 6, 7, 8]

# remove item
list1 = [1, 2, 3, 4, 5]
list1.remove(3)
print(list1)  # Output: [1, 2, 4, 5]

# pop item of list
list1 = [1, 2, 3, 4, 5]
list1.pop(2)
print(list1)  # Output: [1, 2, 4, 5]

# clear list
list1 = [1, 2, 3, 4, 5]
list1.clear()
print(list1)  # Output: []

# sort list
list1 = [5, 2, 3, 4, 1]
list1.sort()
print(list1)  # Output: [1, 2, 3, 4, 5]

# sort list in reverse order
list1 = [5, 2, 3, 4, 1]
list1.sort(reverse=True)
print(list1)  # Output: [5, 4, 3, 2, 1]

# copy list
list1 = [1, 2, 3, 4, 5]
list2 = list1.copy()
print(list2)  # Output: [1, 2, 3, 4, 5]

# count occurrences of an element
list1 = [1, 2, 3, 4, 5, 1]
count = list1.count(1)
print(count)  # Output: 2
```
