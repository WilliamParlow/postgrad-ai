# O que é o Pandas?

O Pandas é uma biblioteca para análise e manipulação de dados para a linguagem Python.

**Característias do Pandas**


*   Estrutura os dados (Dataframe) semelhante a uma planilha com linhas e colunas
*   Leitura e escrita de dados em arquivos CSV, texto, Excel, SQL, etc.
*   Otimização de processamento crítico em linguagem C
*   Facilidade na inserção e exclusão de colunas conferindo mutabilidade nos Dataframe



# Importação do Pandas

```python
import pandas
```

Declaração de um conjunto de dados

```python
dados = {
'cliente': ["Jonas", "Kim", "Lucas"],
'idade': [15, 39, 68]
}
```

#  O que é um DataFrame?

Um DataFrame é uma estrutura de dados bidimensional em que os dados são organizados em linhas e colunas. Os dados podem ser mutáveis, heterogêneo e se parece com uma planilha do Excel.

# Declaração de um dataframe a partir de um conjunto de dados
Para declarar um dataframe, a biblioteca Pandas utiliza a função DataFrame():

```python
df = pandas.DataFrame(dados)
```

Exibição do dataframe

```python
print(df)
```

# Localização de uma linha
A localização de uma linha é realizada por meio do atributo "loc"

```python
print(df.loc[[0]])

print(df.loc[[1]])
```

# Detalhes da linha em outra perspectiva

```python
print(df.loc[0])

print(df.loc[1])
```

# Localizando múltiplas linhas




```python
print(df.loc[[0,2]])
```

# Índices nomeados
Com o argumento index, é possível nomear os itens de um dataframe.

```python
df = pandas.DataFrame(dados, index=["pessoa1", "pessoa2", "pessoa3"])
print(df)
```

# Localizar linhas por índices nomeados
Para localizar linhas por meio dos índices nomeados, deve-se passar o nome do índice para o atributo "loc":

```python
print('------------ loc pessoa2 à 3 -------------')
print(df.loc[["pessoa2", "pessoa3"]])
print("")

print('------------ loc pessoa2 à 1 -------------')
print(df.loc[["pessoa2", "pessoa1"]])
print("")

print('------------ loc pessoa2 à 3 e somente coluna cliente -------------')
print(df.loc["pessoa2":"pessoa3",["cliente"]])
print("")

print('------------ loc todas as pessoas e somente coluna cliente -------------')
print(df.loc[:,["cliente"]])
```
